```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.preprocessing import MinMaxScaler
```

```python
df_=pd.read_excel("english_data.xlsx",sheet_name="Year 2009-2010")
```

```python
df=df_.copy()
df.head()
```

```python
df["TotalPrice"]=df["Quantity"]*df["Price"]
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df=df[~df["Invoice"].str.contains("C",na=False)]
```

```python
df=df[df["Quantity"]>0]
df
```

```python
cltv_c=df.groupby("Customer ID").agg({"Invoice":lambda x:x.nunique(),
                                      "Quantity":"sum",
                                      "TotalPrice":"sum"})

cltv_c.columns=["total_transaction","total_unit","total_price"]
cltv_c
```

```python
cltv_c["avarage_order_value"]=cltv_c["total_price"]/cltv_c["total_transaction"]
cltv_c
```

```python
cltv_c["purchase_frequency"]=cltv_c["total_transaction"]/cltv_c.shape[0]
cltv_c 
```

```python
repeat_rate=cltv_c[cltv_c["total_transaction"] > 1].shape[0]/cltv_c.shape[0]
churn_rate=1-repeat_rate
churn_rate
```

```python
cltv_c["profit_margin"]=cltv_c["total_price"]*0.10
cltv_c
```

```python
cltv_c["customer_value"]=cltv_c["avarage_order_value"] * cltv_c["purchase_frequency"]
cltv_c
```

```python
cltv_c["cltv"]=(cltv_c["customer_value"]/churn_rate )* cltv_c["profit_margin"]
cltv_c
```

```python
cltv_c["segment"]=pd.qcut(cltv_c["cltv"],4,["D","C","B","A"])
cltv_c
```

```python
cltv_c.groupby("segment").agg({"mean"})
```

```python
cltv_c.groupby("segment").agg({"sum"})
```

```python
cltv_c.groupby("segment").agg({"count"})
```
